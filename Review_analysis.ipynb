```python
features=['asin','brand','breadCrumbs','price/value','reviewsCount''seller',
          'seller/name','stars','thumbnailImage','url']
review_features=['date', 'isVerified','productAsin', 'ratingScore', 'reviewDescription',
                 'reviewTitle', 'reviewUrl','variant']
```

```python
needed_col= [x for x in reviews.columns if x in review_features]
needed_col
```

```python
wanted_columns= [x for x in products.columns if x in features]
new_prod= products[wanted_columns]
new_prod.head()
```
